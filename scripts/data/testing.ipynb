```python
def open_file(file):
    l1,l2 = [],[]
    with open(file,'r') as f:
        for i in f:
            x = i.strip().split('|||')
            l1.append(x[0])
            l2.append(x[1])
    return l1, l2
l_src, l_tgt = open_file("/nfs/RESEARCH/avila/WMT2024_NON-REPETITIVE/WMT2024_NON-REPETITIVE_official_data/JIJI_CORPUS_2020/train.txt")
len(l_src), len(l_tgt)
```

## Looking Test Dataset

```python
def open_testset(file1, file2):
    l1,l2=[],[]
    with open(file1, 'r') as f, open(file2, 'r') as g:
        for i,j in zip(f,g):
            l1.append(i.strip())
            l2.append(j.strip())
    return l1,l2
```

```python
import pandas as pd
l_src_dev, l_tgt_dev = open_testset('/nfs/RESEARCH/avila/WMT2024_NON-REPETITIVE/WMT2024_NON-REPETITIVE_official_data/JIJI_CORPUS_2023/wat2023.devtest.tagged.ja','WMT2024_NON-REPETITIVE_official_data/JIJI_CORPUS_2023/wat2023.devtest.tagged.en')
df_test = pd.DataFrame({"source": l_src_dev, "target": l_tgt_dev})
df_test
```

### How many repetitions in Dataset dev?

```python
import re
def count_reps(x):
    pattern=r'<target[^>]*>([^<]*)<\\target>'
    matches = re.findall(pattern, a)
    return len(list(set(matches)))

def brute_reps(x):
    pattern=r'<target[^>]*>([^<]*)<\\target>'
    matches = re.findall(pattern, a)
    return len(matches)

```

```python
df_test['count_s'] = df_test.apply(lambda x: count_reps(x[0]), axis=1)
df_test["count_t"] = df_test.apply(lambda x: count_reps(x[1]), axis=1)
df_test["brute_s"] = df_test.apply(lambda x: brute_reps(x[0]), axis=1)
df_test['diff'] = df_test["brute_s"] - df_test["count_s"]
df_test
```

```python
import torch
import torch.nn.functional as F

# Suponiendo que tienes dos tensores de ejemplo
bs = 32  # Batch size de ejemplo
x = 10   # Dimensión x
y = 15   # Dimensión y
dim = 1024  # Dimensión de características

tensor_a = torch.randn(bs, x, dim)
tensor_b = torch.randn(bs, y, dim)

# Expandir tensor_a para tener la misma dimensión que tensor_b
tensor_a_expanded = tensor_a.unsqueeze(2)  # shape: (bs, x, 1, 1024)
tensor_b_expanded = tensor_b.unsqueeze(1)  # shape: (bs, 1, y, 1024)

# Calcular similitud del coseno a lo largo de la última dimensión (dim=1024)
cosine_sim = F.cosine_similarity(tensor_a_expanded, tensor_b_expanded, dim=-1)

# Hacer un promedio de las similitudes del coseno a lo largo de las dimensiones x e y
average_cosine_sim = cosine_sim.mean(dim=[1, 2])

# Resultado tendrá shape: (bs,)
print(average_cosine_sim.shape)
print(average_cosine_sim)
```
